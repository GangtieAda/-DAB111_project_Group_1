```python
import pandas as pd
import sqlite3
# Connect to the database
conn = sqlite3.connect("cleaned_sample_grades.db") 
cur = conn.cursor()
#Read data
df = pd.read_csv("cleaned_sample_grades.csv")
df.head()
```

```python
# Send data to database
df.to_sql('pd_grades', conn, if_exists='replace', index=False)
```

```python
rows = cur.execute("SELECT * FROM pd_grades").fetchall()
rows
```

```python
#Read from database to dataframe
result = pd.read_sql_query("SELECT * FROM pd_grades", conn)
result
```

```python
#Commit and close
conn.commit()
conn.close()
```
